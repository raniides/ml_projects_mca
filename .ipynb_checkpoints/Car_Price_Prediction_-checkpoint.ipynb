```python
def get_brand_name(car_name):
    car_name = car_name.split(' ')[0]
    return car_name.strip()
```

```python
def clean_data(value):
    value = value.split(' ')[0]
    value = value.strip()
    if value == '':
        value = 0
    return float(value)
```

```python
##Build the Model

#extract the feature and not want drop , will come all in X
#and target column price ,which is dropped will come in y 

X=used_car.drop(columns='selling_price')
y=used_car['selling_price']  #target column
```

```python
X #all features except price
```

```python
y # target column
```

```python
#we will import necessary files all
#train_test_split from sklearn

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y, test_size=0.2)
```

```python
#using linerRegression model
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score #from r2_score we measure regression problems
```

```python
model=LinearRegression()
```

```python
#train model
model.fit(X_train, y_train)
```

```python
pred = model.predict(X_test)
```

```python
pred
```

```python
accuracy = r2_score(y_test, pred)

print("Accuracy (R-squared score):", accuracy)
```

```python
import pickle
pickle.dump(model,open('LinearRegressionModel.pkl','wb'))
```

```python
if accuracy >= 0.7:
    print("The model is doing a good job explaining the variation in prices.")
elif 0.5 <= accuracy < 0.7:
    print("The model is decent but may be improved.")
else:
    print("The model needs improvement as it doesn't explain the variation well.")
```

```python
# Get user input for feature values
name = input("Enter the vehicle name (e.g., Maruti, Hyundai, etc.): ")
year = int(input("Enter the vehicle purchased year: "))
fuel_type = input("Enter the fuel type (Diesel, Petrol, LPG, CNG, Electric): ")
seller_type = input("Enter the seller type (Individual, Dealer, Trustmark Dealer): ")
transmission = input("Enter the transmission type (Manual, Automatic): ")
owner = input("Enter the owner type (First Owner, Second Owner, Third Owner, Fourth & Above Owner, Test Drive Car): ")
km_driven = float(input("Enter the kilometers driven: "))

# Map user input to numerical categories based on the provided mappings
name = {
    'Maruti': 1, 'Hyundai': 2, 'Datsun': 3, 'Honda': 4, 'Tata': 5, 'Chevrolet': 6,
    'Toyota': 7, 'Jaguar': 8, 'Mercedes-Benz': 9, 'Audi': 10, 'Skoda': 11, 'Jeep': 12,
    'BMW': 13, 'Mahindra': 14, 'Ford': 15, 'Nissan': 16, 'Renault': 17, 'Fiat': 18,
    'Volkswagen': 19, 'Volvo': 20, 'Mitsubishi': 21, 'Land': 22, 'Daewoo': 23, 'MG': 24,
    'Force': 25, 'Isuzu': 26, 'OpelCorsa': 27, 'Ambassador': 28, 'Kia': 29
}.get(name)

transmission = {'Manual': 1, 'Automatic': 2}.get(transmission)
seller_type = {'Individual': 1, 'Dealer': 2, 'Trustmark Dealer': 3}.get(seller_type)
fuel_type = {'Diesel': 1, 'Petrol': 2, 'LPG': 3, 'CNG': 4, 'Electric': 5}.get(fuel_type)
owner = {'First Owner': 1, 'Second Owner': 2, 'Third Owner': 3, 'Fourth & Above Owner': 4, 'Test Drive Car': 5}.get(owner)


# Make a prediction based on user input
user_input = pd.DataFrame([[name, year, km_driven, fuel_type, seller_type,transmission
                           ,owner]],columns=['name', 'year', 'km_driven', 'fuel_type',
                            'seller_type','transmission','owner'])

predicted_price = model.predict(user_input)

print(f"Predicted Price for the given input: ${predicted_price[0]:.2f}")
```
